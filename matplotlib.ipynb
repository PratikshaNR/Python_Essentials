## **Matplotlib Library**

Free and open-source plotting library in Python.

Used to create graphs and charts.

Works with NumPy (arrays) and Pandas (DataFrames).

Can make many plots: line, bar, scatter, pie, histogram, box, etc.

Customizable: colors, labels, titles, legends, grids.

Has two ways to use:

pyplot – quick and easy plots.

Object-oriented – for detailed/custom plots.

Can save graphs as images (PNG, JPG, PDF).


- Basic plots are :
  - Scatter Plot :  
    - (1) Understand how data is distributed in two dimensions.
    - (2) Identify the relationship/correlation between two numerical features.
    - (3) Detect clusters, patterns, or anomalies.
    - Axes: Both X and Y are numerical.
    - Extra: Can be color-coded or sized by a third variable for multivariate analysis.


  - Bar plot :  
     - (1) Compare categories (frequency or values).
     - (2) Can be used for bivariate analysis (categorical vs numerical).
     - (3) Shows differences clearly for discrete data.
     - Axes:X-axis → Categorical, Y-axis → Numerical (or frequency/count).
     - Extra: Horizontal bar plots are good for long category names.


  - Line plot:  
    -  (1) how trends over time (time-series analysis).
    -  (2) Track changes between data points in a sequence.
    - Axes:X-axis → Time or ordered sequence, Y-axis → Numerical values.
    - Extra: Multiple lines can compare trends across categories.


  - Pie chart:  
     - (1) Show distribution as proportions or percentages.
     - (2) Compare parts of a whole.
     - Features:Contains labels and values.
        Total should sum to 100%.
     - Extra: Works best with fewer categories (≤6) for clarity.



  - Histogram: 
    - Display distribution of a numerical variable.
    - Group data into bins to see frequency.
    - Extra: Good for checking skewness, spread, and modality.

                
  - Density plot: 
    - Show the probability density of a continuous variable.
    - Smoother alternative to a histogram.
    - Extra: Useful for comparing multiple distributions on the same plot.



  - Box plot/Wiskers plot: 
    - Identify spread of data and potential outliers.
    - Shows median, quartiles (Q1, Q3), and whiskers.
    - Extra: Can be drawn vertically or horizontally; useful for comparing groups.


  - Combine plot: 
    - Combine multiple plots in one figure for comparison.
    - Useful in exploratory data analysis (EDA).
    - Extra:Can mix different plot types (scatter + histogram, etc.).
    - Allows customization like shared axes, colors, and annotations.

## **Violin Plot**
It is  a data visulisation combines features of box plot and kernal density plot.
provides insights into the distribution of a dataset by showing both summary satistics and the data's density shape.
Handling outliers.
uses for univariate analysis.

```python
sns.violinplot(data=df,x='tip',palette='winter',inner_kws=dict(box_width=15,whis_width=2),split=True)
plt.title('Distribution of tip',fontsize=15,fontweight='bold',fontname='arial',loc='left',pad=10)
plt.show()
```
